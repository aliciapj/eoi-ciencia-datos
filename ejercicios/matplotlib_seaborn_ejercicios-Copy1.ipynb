## The Data

We will be working with a famous titanic data set for these exercises. Later on in the Machine Learning section of the course, we will revisit this data, and use it to predict survival rates of passengers. For now, we'll just focus on the visualization of the data with seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Ejericicio 1: Visualizaciones básicas

- Visualiza cuántas personas vivieron vs cuántas murieron en el accidente

```python
ax = sns.countplot(x='survived', data=titanic)
ax.set(title = 'Supervivientes totales', xlabel = 'Supervivientes', ylabel='Total')
plt.show()
```

- Comparativa de pasajeros que vivieron o murieron dependiendo de su clase (`pclass`)

```python
ax = sns.countplot(x='pclass', hue='survived', data=titanic)
ax.set(title = 'Supervivientes según la clase del pasaje', xlabel = 'Clase', ylabel='Total')
plt.show()
```

- Comparativa de pasajeros que murieron dependiendo de su género (`sex`)

```python
ax = sns.countplot(x='sex', hue='survived', data=titanic)
ax.set(title = 'Supervivientes según el género', xlabel = 'Género', ylabel='Total')
plt.show()
```

- Comparativa de pasajeros que murieron o sobrevivieron según la puerta de embarque (`embarked`)

```python
ax = sns.countplot(x='embarked', hue='survived', data=titanic)
ax.set(title = 'Supervivientes según la puerta de embarque', xlabel = 'Puerta', ylabel='Total')
plt.show()
```

- Comparativa de pasajeros que murieron dependiendo del número de personas que les acompañaban (`sibsp`)

```python
ax = sns.countplot(x='sibsp', hue='survived', data=titanic)
ax.set(title = 'Supervivientes según el número de acompañantes', xlabel = '# Acompañantes', ylabel='Total')
plt.show()
```

- Comparativa de pasajeros que murieron o sobrevivieron según el rango de edad (`age`)
> Pista 1: prueba a hacer intervalos de edad con `pd.cut`

```python
ax = sns.countplot(x='age', hue='survived', data=titanic)
ax.set(title = 'Supervivientes según la edad', xlabel = 'Edad', ylabel='Total')
plt.show()
```

```python
interval = (0,18,35,60,120)
categories = ['Children','Teens','Adult', 'Old']
titanic['age_cats'] = pd.cut(titanic.age, interval, labels=categories)

ax = sns.countplot(x = 'age_cats',  data =titanic, hue = 'survived')

ax.set(title="Supervivientes según el rango de edad", xlabel='Rango de edad', ylabel='Total')

plt.show()
```

- Comparativa de pasajeros que murieron según el precio del pasaje (`fare`)
> Prueba con un `FacetGrid`

```python
fare = sns.FacetGrid(titanic, hue="survived", aspect=2)
fare.map(sns.kdeplot, 'fare', shade= True)
fare.set(xlim=(0, 200))
fare.add_legend()
```
